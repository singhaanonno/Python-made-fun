```python
from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
import matplotlib.pyplot as plt #to display our wordcloud
from PIL import Image #to load our image
import numpy as np #to get the color of our image
```

### General simple plot

```python
text=open('Untitled.txt','r').read()

wc=WordCloud()
wc.generate(text)

plt.imshow(wc,interpolation='bilinear')
plt.show()
```

### Changing the Background Color

```python
# content related
text=open('Untitled.txt','r').read()

# Appearence-related
wc=WordCloud(background_color='white') # Now lets change the background color 
wc.generate(text)

#plotting
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')  # we wanna get rid of the axis
plt.show()
```

### Giving the maximum font size

```python
# content related
text=open('Untitled.txt','r').read()

# Appearence-related
wc=WordCloud(background_color='white',max_font_size=50)  # we can control our font size as well
wc.generate(text)

#plotting
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')  
plt.show()
```

### Setting the limit of Maximum Words

```python
# content related
text=open('Untitled.txt','r').read()

# Appearence-related
wc=WordCloud(background_color='white',max_font_size=100,max_words=10)  
wc.generate(text)

#plotting
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')  
plt.show()
```

### WE CAN ALSO GET RID OF SOME WORDS

```python
# content related
text=open('Untitled.txt','r').read()
stopwords=set(STOPWORDS)

# Appearence-related
wc=WordCloud(background_color='gray',max_font_size=100,max_words=30,stopwords=stopwords)  
wc.generate(text)

#plotting
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')  
plt.show()
```

### Changing the size of the plot

```python
# content related
text=open('Untitled.txt','r').read()
stopwords=set(STOPWORDS)

# Appearence-related
wc=WordCloud(background_color='gray',max_font_size=100,max_words=30,stopwords=stopwords,height=500,width=2000)  
wc.generate(text)

#plotting
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')  
plt.show()
```
